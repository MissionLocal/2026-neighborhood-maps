```python
import geopandas as gdp
import pandas as pd
```

```python
# read neighborhoods geojson

neighborhoods = gdp.read_file("neighborhoods.geojson")
```

```python
neighborhoods
```

Tenderloin:

```python
tenderloin = neighborhoods[neighborhoods["nhood"] == "Tenderloin"]
```

```python
tenderloin.to_file("../docs/tenderloin/tenderloin.geojson", driver="GeoJSON")
```

Bayview: 

```python
bayview_hunters_point = neighborhoods[neighborhoods["nhood"] == "Bayview Hunters Point"]
```

```python
bayview_hunters_point.to_file("../docs/bayview/bayview.geojson", driver="GeoJSON")
```

Excelsior: 

```python
excelsior = neighborhoods[neighborhoods["nhood"] == "Excelsior"]
```

```python
excelsior.to_file("../docs/excelsior/excelsior.geojson", driver="GeoJSON")
```

Richmond:

```python
two_neighborhoods = neighborhoods[
    neighborhoods["nhood"].isin(["Inner Richmond", "Inner Richmond"])
]

merged = two_neighborhoods.dissolve(by="nhood")

merged.to_file("../docs/richmond/richmond.geojson", driver="GeoJSON")
```

Sunset:

```python
two_neighborhoods = neighborhoods[
    neighborhoods["nhood"].isin(["Inner Sunset", "Outer Sunset"])
]

merged = two_neighborhoods.dissolve(by="nhood")

merged.to_file("../docs/sunset/sunset.geojson", driver="GeoJSON")
```
