```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('./ACSDP5Y2024.DP05-Data.csv')
```

```python
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# print all column names
print(df.columns.tolist())
```

```python
# rename columns
df = df.rename(columns={
    'Estimate!!SEX AND AGE!!Total population!!Under 5 years': 'Under 5 years',
    'Estimate!!SEX AND AGE!!Total population!!5 to 9 years': '5 to 9 years',
    'Estimate!!SEX AND AGE!!Total population!!10 to 14 years': '10 to 14 years',
    'Estimate!!SEX AND AGE!!Total population!!15 to 19 years': '15 to 19 years',
    'Estimate!!SEX AND AGE!!Total population!!20 to 24 years': '20 to 24 years',
    'Estimate!!SEX AND AGE!!Total population!!25 to 34 years': '25 to 34 years',
    'Estimate!!SEX AND AGE!!Total population!!35 to 44 years': '35 to 44 years',
    'Estimate!!SEX AND AGE!!Total population!!45 to 54 years': '45 to 54 years',
    'Estimate!!SEX AND AGE!!Total population!!55 to 59 years': '55 to 59 years',
    'Estimate!!SEX AND AGE!!Total population!!60 to 64 years': '60 to 64 years',
    'Estimate!!SEX AND AGE!!Total population!!65 to 74 years': '65 to 74 years',
    'Estimate!!SEX AND AGE!!Total population!!75 to 84 years': '75 to 84 years',
    'Estimate!!SEX AND AGE!!Total population!!85 years and over': '85 years and over',
})
# only keep the columns we want 
df = df[['Area', 'Under 5 years', '5 to 9 years', '10 to 14 years', '15 to 19 years', 
         '20 to 24 years', '25 to 34 years', '35 to 44 years', 
         '45 to 54 years', '55 to 59 years', '60 to 64 years', 
         '65 to 74 years', '75 to 84 years', '85 years and over']]
df.head(3)
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df
```

```python
# combine age groups
df['5 to 19 years'] = df[['5 to 9 years', '10 to 14 years', '15 to 19 years']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['20 to 34 years'] = df[['20 to 24 years', '25 to 34 years']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['35 to 59 years'] = df[['35 to 44 years', '45 to 54 years', '55 to 59 years']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['60 to 74 years'] = df[['60 to 64 years', '65 to 74 years']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['75 plus'] = df[['75 to 84 years', '85 years and over']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df.head(3)
# only keep the columns we want
df = df[['Area', 'Under 5 years', '5 to 19 years', '20 to 34 years', 
         '35 to 59 years', '60 to 74 years', '75 plus']]
df.head(3)
```

```python
# rename the columns to make them more readable
df = df.rename(columns={
    'Under 5 years': 'Under 5',
    '5 to 19 years': '5 to 19',
    '20 to 34 years': '20 to 34',
    '35 to 59 years': '35 to 59',
    '60 to 74 years': '60 to 74',
    '75 plus': '75 plus'
})
df.head(3)
```

```python
# get total
df['Total'] = df[['Under 5', '5 to 19', '20 to 34', '35 to 59', '60 to 74', '75 plus']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df.head()
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Under 5', '5 to 19', '20 to 34', '35 to 59', '60 to 74', '75 plus', 'Total']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'under 5': df.set_index('Area')['Under 5'].to_dict(),
    '5 to 19': df.set_index('Area')['5 to 19'].to_dict(), 
    '20 to 34': df.set_index('Area')['20 to 34'].to_dict(),
    '35 to 59': df.set_index('Area')['35 to 59'].to_dict(),
    '60 to 74': df.set_index('Area')['60 to 74'].to_dict(),
    '75 plus': df.set_index('Area')['60 to 74'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('../../data/age.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'age.json'")
```
