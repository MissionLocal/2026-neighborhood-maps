```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('./ACSDT5Y2024.B03002-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# rename columns
df = df.rename(columns={
    'Estimate!!Total:': 'Citywide',
    'Estimate!!Total:!!Not Hispanic or Latino:': 'Citywide Non-Hispanic or Latino',
    'Estimate!!Total:!!Not Hispanic or Latino:!!White alone': 'White',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Black or African American alone': 'Black',
    'Estimate!!Total:!!Not Hispanic or Latino:!!American Indian and Alaska Native alone': 'Other1',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Asian alone': 'Asian',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Native Hawaiian and Other Pacific Islander alone': 'Other2',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Some other race alone': 'Other3',
    'Estimate!!Total:!!Not Hispanic or Latino:!!Two or more races:': 'Other4',
    'Estimate!!Total:!!Hispanic or Latino:': 'Hispanic or Latino'})
# only keep the columns we want
df = df[['Area', 'Citywide', 'Citywide Non-Hispanic or Latino', 'White', 'Black', 'Other1', 'Asian', 'Other2', 'Other3', 'Other4', 'Hispanic or Latino']]
df.head(3)
```

```python
# convert to numeric and sum in one step
df['Other'] = df[['Other1', 'Other2', 'Other3', 'Other4']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df.head(3)
```

```python
# only keep the columns we want
df = df[['Area', 'Citywide', 'White', 'Black', 'Asian', 'Hispanic or Latino', 'Other']]
df.head(3)
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Citywide', 'White', 'Black', 'Asian', 'Hispanic or Latino', 'Other']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'white': df.set_index('Area')['White'].to_dict(),
    'asian': df.set_index('Area')['Asian'].to_dict(), 
    'hispanic': df.set_index('Area')['Hispanic or Latino'].to_dict(),
    'black': df.set_index('Area')['Black'].to_dict(),
    'other': df.set_index('Area')['Other'].to_dict(),
    'total': df.set_index('Area')['Citywide'].to_dict()
}

# convert to JSON and save to file
import json

with open('../../data/race.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'race.json'")
```
