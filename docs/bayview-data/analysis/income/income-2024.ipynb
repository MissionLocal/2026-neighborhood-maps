```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('./ACSDT5Y2024.B19001-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df.head(3)
```

```python
# get rid of Estimate!!Total:!! in column names
df.columns = df.columns.str.replace('Estimate!!Total:!!', '', regex=True)
df.head(3)
```

```python
# combine a few columns
df['Under $20k'] = df[['Less than $10,000', '$10,000 to $14,999', '$15,000 to $19,999']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['$20-60k'] = df[['$20,000 to $24,999', '$25,000 to $29,999', '$30,000 to $34,999', '$35,000 to $39,999', '$40,000 to $44,999', '$45,000 to $49,999', '$50,000 to $59,999']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['$60-100k'] = df[['$60,000 to $74,999', '$75,000 to $99,999']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['$100-150k'] = df[['$100,000 to $124,999', '$125,000 to $149,999']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['$150-200k'] = df[['$150,000 to $199,999']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['$200k plus'] = df[['$200,000 or more']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
# drop the original columns
columns_to_drop = [
    'Less than $10,000', '$10,000 to $14,999', '$15,000 to $19,999',
    '$20,000 to $24,999', '$25,000 to $29,999', '$30,000 to $34,999',
    '$35,000 to $39,999', '$40,000 to $44,999', '$45,000 to $49,999',
    '$50,000 to $59,999', '$60,000 to $74,999', '$75,000 to $99,999',
    '$100,000 to $124,999', '$125,000 to $149,999', '$150,000 to $199,999',
    '$200,000 or more'
]
df = df.drop(columns=columns_to_drop)
df.head(3)
```

```python
# change estimate total column name to Total
df = df.rename(columns={'Estimate!!Total:': 'Total'})
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total', 'Under $20k', '$20-60k', '$60-100k', '$100-150k', '$150-200k', '$200k plus']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'under $20k': df.set_index('Area')['Under $20k'].to_dict(),
    '$20-60k': df.set_index('Area')['$20-60k'].to_dict(), 
    '$60-100k': df.set_index('Area')['$60-100k'].to_dict(),
    '$100-150k': df.set_index('Area')['$100-150k'].to_dict(),
    '$150-200k': df.set_index('Area')['$150-200k'].to_dict(),
    '$200k plus': df.set_index('Area')['$200k plus'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('../../data/income.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'income.json'")
```
