```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('/Users/agneslu/Desktop/neighborhood-data/data-new/vehicle/ACSDT5Y2023.B25044-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df.head(3)
```

```python
# get rid of Estimate!!Total:!! in column names
df.columns = df.columns.str.replace('Estimate!!Total:!!', '', regex=True)
df.head(3)
```

```python
# combine a few columns
df['No vehicle'] = df[['Owner occupied:!!No vehicle available', 'Renter occupied:!!No vehicle available']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['One vehicle'] = df[['Owner occupied:!!1 vehicle available', 'Renter occupied:!!1 vehicle available']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Two vehicles'] = df[['Owner occupied:!!2 vehicles available', 'Renter occupied:!!2 vehicles available']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Three or more vehicles'] = df[['Owner occupied:!!3 vehicles available', 'Owner occupied:!!4 vehicles available', 'Owner occupied:!!5 or more vehicles available', 'Renter occupied:!!3 vehicles available', 'Renter occupied:!!4 vehicles available', 'Renter occupied:!!5 or more vehicles available']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
# only keep the columns we need
columns_to_keep = ['Area', 'Estimate!!Total:', 'No vehicle', 'One vehicle', 'Two vehicles', 'Three or more vehicles']
df = df[columns_to_keep]
df.head(3)
```

```python
# change estimate total column name to Total
df = df.rename(columns={'Estimate!!Total:': 'Total'})
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total', 'No vehicle', 'One vehicle', 'Two vehicles', 'Three or more vehicles']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'no vehicle': df.set_index('Area')['No vehicle'].to_dict(),
    'one vehicle': df.set_index('Area')['One vehicle'].to_dict(), 
    'two vehicles': df.set_index('Area')['Two vehicles'].to_dict(),
    'three or more vehicles': df.set_index('Area')['Three or more vehicles'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('/Users/agneslu/Desktop/neighborhood-data/data-new/vehicles.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'vehicles.json'")
```
