```python
import pandas as pd
```

```python
# import data, using the dataset I used for age
df = pd.read_csv('/Users/agneslu/Desktop/neighborhood-data/analysis/age/ACSDP5Y2023.DP05-Data.csv')
```

```python
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# keep area and total population columns
df = df[['Area', 'Estimate!!SEX AND AGE!!Total population']]
df.head(3)
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df
```

```python
# change the total population column name
df = df.rename(columns={'Estimate!!SEX AND AGE!!Total population': 'Total Population'})
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total Population']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'value': df.set_index('Area')['Total Population'].to_dict(),
}

# convert to JSON and save to file
import json

with open('/Users/agneslu/Desktop/neighborhood-data/data/population.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'population.json'")
```

```python
# export the final dataframe to CSV
df.to_csv('/Users/agneslu/Desktop/neighborhood-data/analysis/crime/population.csv', index=False)
```
