```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('./ACSDT5Y2024.B15003-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
df.head(3)
```

```python
# drop the first row again
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# get rid of Estimate!!Total: in column names
df.columns = df.columns.str.replace('Estimate!!Total:!!', '', regex=False)
df.head()
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df.head(3)
```

```python
# Less than high school
df['Less_than_High_School'] = df[['No schooling completed', 'Nursery school', 'Kindergarten', '1st grade', '2nd grade', '3rd grade', '4th grade', '5th grade', '7th grade', '8th grade', '9th grade', '10th grade', '11th grade', '12th grade, no diploma']].apply(pd.to_numeric, errors='coerce').sum(axis=1) 

# High school or equivalent
df['High_School_or_Equivalent'] = df[['Regular high school diploma', 'GED or alternative credential']].apply(pd.to_numeric, errors='coerce').sum(axis=1) 

# Some college, no degree
df['Some_College_No_Degree'] = df[['Some college, less than 1 year', 'Some college, 1 or more years, no degree']].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Bachelor's or associate's degree
df['Bachelors_or_Associates'] = df[['Associate\'s degree', 'Bachelor\'s degree']].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Masters's or professional's degree
df['Masters_or_Professional'] = df[['Master\'s degree', 'Professional school degree']].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Doctorate degree
df['Doctorate'] = df[['Doctorate degree']].apply(pd.to_numeric, errors='coerce').sum(axis=1) 

df = df[['Area', 'Estimate!!Total:', 'Less_than_High_School', 'High_School_or_Equivalent', 'Some_College_No_Degree', 'Bachelors_or_Associates', 'Masters_or_Professional', 'Doctorate']]
df.head(3)
```

```python
# change Estimate!!Total: to Total
df = df.rename(columns={'Estimate!!Total:': 'Total'})
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total', 'Less_than_High_School', 'High_School_or_Equivalent', 'Some_College_No_Degree', 'Bachelors_or_Associates', 'Masters_or_Professional', 'Doctorate']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'less than high school': df.set_index('Area')['Less_than_High_School'].to_dict(),
    'high school or equivalent': df.set_index('Area')['High_School_or_Equivalent'].to_dict(), 
    'some college, no degree': df.set_index('Area')['Some_College_No_Degree'].to_dict(),
    'bachelor\'s or associate\'s degree': df.set_index('Area')['Bachelors_or_Associates'].to_dict(),
    'masters\'s or professional\'s degree': df.set_index('Area')['Masters_or_Professional'].to_dict(),
    'doctorate': df.set_index('Area')['Doctorate'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('../../data/education.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'education.json'")
```
